```python
import pandas as pd
import numpy as np 
import  matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings("ignore")
```

```python
# For AND Gate

x = np.array([[0,0], [0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
p.fit(x,y)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x,y , clf = p, legend = 2)
plt.title("AND GATE")
plt.show()
```

```python
# For OR Gate

x = np.array([[0,0], [0,1],[1,0],[1,1]])
y = np.array([0,1,1,1])

p = Perceptron()
p.fit(x,y)

plot_decision_regions(x,y , clf = p, legend = 2)
plt.title("OR GATE")
plt.show()
```

```python
# for XOR Gate 

x = np.array([[0,0], [0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

p = Perceptron()
p.fit(x,y)

plot_decision_regions(x,y , clf = p, legend = 2)
plt.title("XOR GATE")
plt.show()
```
